# align CC score to SNPs

```r
# ---- Set the following paths to match your environment ----
wd_path  = "/path/to/shell_script"        # directory of shell script
data_path  = "/path/to/data"              # directory for data
save_path  = "/path/to/output"            # directory for output
TF_ChIP_path = "/path/to/ChIP_seq_data"   # directory for ChIP-seq peak data
1000G_EUR_plink_path = "/path/to/1000G_Phase3_EUR_plinkfiles" # directory for 1000G_Phase3_EUR_plinkfile
# -----------------------------------------------------------

cd ${wd_path}
mkdir log
cd log


for CC in CC1_rotation CC2_rotation CC3_rotation CC4_rotation CC5_rotation CC6_rotation CC7_rotation CC8_rotation CC9_rotation CC10_rotation
do
qsub -pe def_slot 1 \
   -l s_vmem=4G \
   -cwd \
    ${wd}/align_CCscore_to_SNPs.sh ${CC} ${data_path} ${save_path} ${TF_ChIP_path} ${1000G_EUR_plink_path}
done
```
