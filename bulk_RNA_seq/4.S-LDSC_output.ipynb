```r
library(ggplot2) 

set.seed(1)
library(fs)

library(data.table) 
library(dplyr)
library(tidyverse)
library(psych)
```

```r
data_path=<path_to_data>
dir_path=paste0(data_path, "/CC_10percent_SNP_noexp_2/sldsc/EUR")
filelist=dir(dir_path, pattern="_sum_SNP")


enrichment_allsumstats_df=data.frame(matrix(rep(NA, 11), ncol=11))[0,]

for (file in filelist){
    sumstats_list=dir(paste0(dir_path, "/", file), pattern=".results")
    for (sumstats in sumstats_list){
        df=read.table(paste0(dir_path, "/", file, "/", sumstats), header=1)
        annotation_df=df[1, 2:10]
        metadata=strsplit(file, "_")
        new_df=data.frame(sumstatsname=sumstats,CC_top_bottom=paste0(metadata[[1]][3],"_", metadata[[1]][5]), annotation_df)
        enrichment_allsumstats_df=rbind(enrichment_allsumstats_df, new_df)
    }
}

```

```r
join_df=enrichment_allsumstats_df
```

```r
join_df$minuslog10p=-log10(pnorm(-join_df$Coefficient_z.score,0,1))

join_df$CC_top_bottom=factor(join_df$CC_top_bottom, levels=c("CC1_top","CC1_bottom", "CC2_top","CC2_bottom", "CC3_top", "CC3_bottom","CC4_top","CC4_bottom", "CC5_top","CC5_bottom", "CC6_top","CC6_bottom", "CC7_top","CC7_bottom", "CC8_top","CC8_bottom", "CC9_top", "CC9_bottom","CC10_top", "CC10_bottom"))

```
